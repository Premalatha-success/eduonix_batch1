```python
### Univariate Distribution of numerical attribute(int and float datatypes) - Single attribute

sns.distplot(df["highway_mpg"])
plt.show()
```

```python
### title->Distribution of wheel base
```

```python
sns.distplot(df['wheel_base'],color="brown")
plt.title("Distribution of wheel_base",color="blue",fontsize=20)
plt.xlabel("wheel_base",color="red",fontsize=15,loc="right")
plt.ylabel("Density",color="red",fontsize=15,loc="top")
plt.show()
```

```python
sns.distplot(df['horsepower'],color="green")
plt.show()
```

```python
sns.distplot(df['wheel_base'],kde=False,color="brown")
plt.title("Distribution of wheel_base",color="blue",fontsize=20)
plt.xlabel("wheel_base",color="red",fontsize=15,loc="right")
plt.ylabel("Density",color="red",fontsize=15,loc="top")
plt.show()
```

```python
### Univariate Distribution for categorical(Object)
sns.countplot(x="fuel_system",data=df,color="k")
plt.title("Count of fuel_system",color="brown")
plt.xlabel("fuel_system")
plt.ylabel("frequency")
plt.show()
```

```python
df["fuel_system"].value_counts()
```

```python
### Univariate Distribution for categorical(Object)
sns.countplot(x="fuel_system",data=df,palette=["blue","red","green","purple","orange","black","yellow","pink"])
plt.title("Count of fuel_system",color="brown")
plt.xlabel("fuel_system")
plt.ylabel("frequency")
plt.show()
```

```python
### Bivariate(Two attributes-> Both numerical)
sns.pairplot(df)## corelation
plt.show()
```

```python
sns.relplot(x="bore",y="stroke",data=df)
plt.show()
```

```python
sns.jointplot(x="bore",y="stroke",data=df)
plt.show()
```

```python
### One categorical and one numerical
sns.catplot(x="fuel_type",y="bore",data=df)
plt.show()
```

```python
sns.catplot(x="fuel_type",y="bore",kind="swarm",data=df,palette="Set2")
plt.show()
```

```python
sns.catplot(x="fuel_type",y="bore",kind="violin",data=df,palette="coolwarm")
plt.show()
```

```python
sns.catplot(x="fuel_type",y="bore",kind="box",data=df,palette="Set2")
plt.show()
```

```python
### Two categorical
sns.countplot(x="fuel_type",hue="engine_type",data=df)
plt.show()
```

```python
#### Multivariate(More than 2 attributes)-> two numerical and one categorical
sns.relplot(x="bore",y="stroke",hue="fuel_type",data=df)
plt.show()
```

```python
sns.relplot(x="curb_weight",y="bore",hue="engine_size",data=df)
plt.show()
```

```python
sns.relplot(x='highway_mpg',y='city_mpg',hue='body_style',data=df)
plt.show()
```

```python
sns.catplot(x="fuel_type",y="bore",hue="fuel_type",data=df)
plt.show()
```
